# Gradient Descent from scrach

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
n_samples = 100

x = np.linspace(0, 10, n_samples).reshape(-1, 1)

real_w = 2.0
real_b = -3.0

y = real_w * x + real_b
plt.plot(x, y);
```

```python
import numpy as np

np.random.seed(42)
n_samples = 100
lr = 0.001
tolerance = 0.00001
num_iters = 10000
losses = []

w = np.random.randn()
b = np.random.randn()

prev_loss = float("inf")  # Initialize previous loss to infinity for comparison

for iteration in range(num_iters):
    y_pred = w * x + b
    loss = np.sum((y - y_pred) ** 2) / n_samples  # Scale the loss by n_samples
    losses.append(loss)

    if abs(prev_loss - loss) < tolerance:  # Check for convergence
        print(f"After {iteration} iterations, converged!")
        break
    prev_loss = loss  # Update previous loss

    w_slope = (-2) * np.sum(x * (y - y_pred)) / n_samples  # Scale the gradient
    b_slope = (-2) * np.sum(y - y_pred) / n_samples  # Scale the gradient

    w = w - (w_slope * lr)
    b = b - (b_slope * lr)

print(f"real weight = {real_w:.3f}, read bias = {real_b:.3f}")
print(f"pred weight = {w:.3f}, pred_bias = {b:.3f}")
```
